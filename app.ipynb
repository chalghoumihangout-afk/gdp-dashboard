```python
import streamlit as st
import numpy as np
import joblib

# ==========================
# Chargement du mod√®le
# ==========================
# model.pkl doit √™tre dans le m√™me dossier que app.py
model = joblib.load("model.pkl")

# ==========================
# Interface Streamlit
# ==========================
st.title("üîÆ Application de Pr√©diction ML")

st.write("""
Cette application utilise un mod√®le de Machine Learning
(ici une r√©gression lin√©aire) pour faire une pr√©diction
√† partir d'une valeur num√©rique.
""")

# Champ de saisie utilisateur
x = st.number_input(
    "Entrez une valeur :",
    min_value=0.0,
    step=1.0
)

# ==========================
# Pr√©diction
# ==========================
if st.button("Pr√©dire"):
    # Le mod√®le attend un tableau 2D ‚Üí [[x]]
    x_input = np.array([[x]])
    prediction = model.predict(x_input)

    st.success(f"üìå R√©sultat de la pr√©diction : **{prediction[0]:.2f}**")
```
